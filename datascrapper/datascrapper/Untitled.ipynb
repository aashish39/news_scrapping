```python
import json

# Open the JSON file
with open('data.json', 'r') as file:  # Replace 'data.json' with your file path
    data = json.load(file)  # Parse JSON data
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data)
```

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.prnewswire.com/news-releases/union-presidents-of-apwu-nalc-and-nrlca-to-speak-at-a-national-press-club-newsmaker-on-mar-25-302407576.html"  # Replace with the target website URL
response = requests.get(url)  # Send an HTTP GET request
html_content = response.text  # Extract the HTML content
```

```python
soup = BeautifulSoup(html_content, 'html.parser')
```

```python
plain_text = soup.get_text()
# print(plain_text)
```

```python
data = []
for x in df['links'][:10]:
    response = requests.get("https://www.prnewswire.com" + x)
    html_content = response.text
    soup = BeautifulSoup(html_content, 'html.parser')
    data.append(soup.get_text())
```
