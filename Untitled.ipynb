```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.prnewswire.com/news-releases/oros-crested-butte-the-pinnacle-of-alpine-living-302392847.html'  # Replace with the URL you want to scrape
response = requests.get(url)

if response.status_code == 200:
    html_content = response.text
else:
    print("Failed to fetch the webpage")
```

```python
soup = BeautifulSoup(html_content, 'html.parser')  # You can also use 'lxml' if installed
```

```python
webpage_text = soup.get_text()
```

```python
cleaned_text = ' '.join(webpage_text.split())
print(cleaned_text)
```
